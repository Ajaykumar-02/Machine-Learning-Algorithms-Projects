```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('species', axis=1)
y = df['species']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier(n_neighbors=13)
```

```python
model_knn.fit(X_train, y_train)
```

```python
model_knn.score(X_test, y_test)
```

```python
from sklearn.svm import SVC
```

```python
model_svm = SVC(gamma='auto')
```

```python
model_svm.fit(X_train, y_train)
```

```python
model_svm.score(X_test, y_test)
```

```python
# now lets use the Random search CV
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# We Use SVM Model
```

```python
classifier_r = RandomizedSearchCV((model_svm),{
    'C':[1,10,20,30],
    'kernel':['rbf','linear'],
},n_iter=5,cv=5,return_train_score=False)
```

```python
classifier_r.fit(X,y)
```

```python
classifier_r.cv_results_
```

```python
results = pd.DataFrame(classifier_r.cv_results_)
```

```python
results
```

```python
results[['param_C','param_kernel','mean_test_score']]
```

```python
# We use KNN Model
```

```python
classifier_2r = RandomizedSearchCV((model_knn),{
    'n_neighbors':[3,5,7,9,11],
     'weights':['uniform','distance'],
    'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'],

},n_iter=10,cv = 5, return_train_score=False)
```

```python
classifier_2r.fit(X,y)
```

```python
classifier_2r.cv_results_
```

```python
results_knn = pd.DataFrame(classifier_2r.cv_results_)
```
